## *Histogram by Groups* dari DataFrame

```python
# Membuat Histogram by Groups
fig = px.histogram(data_df,
                   x = 'Nilai',
                   color = 'Mata Pelajaran',
                   barmode = 'overlay',
                   title = "Histogram by Groups dari DataFrame")   # Judul Gambar

fig.show()
```

### Mengubah Warna pada *Histogram by Groups*

```python
# Membuat Histogram by Groups
fig = px.histogram(data_df,
                   x = 'Nilai',
                   color = 'Mata Pelajaran',
                   barmode = 'overlay',
                   color_discrete_sequence = ['green',             # Mengubah Warna Histogram
                                              'orange'],
                   title = "Histogram by Groups dari DataFrame")   # Judul Gambar

fig.show()
```

### Mengubah Banyak *Bin* pada *Histogram by Groups*

```python
# Membuat Histogram by Groups
fig = px.histogram(data_df,
                   x = 'Nilai',
                   color = 'Mata Pelajaran',
                   barmode = 'overlay',
                   nbins = 4,                                      # Mengatur Banyak Bin
                   title = "Histogram by Groups dari DataFrame")   # Judul Gambar

fig.show()
```
